# basic usage

```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline --no-import-all
```

```python
x = np.arange(-3, 3, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.random.randn(30)
y = np.random.randn(30)
plt.plot(x, y, 'ro')
```
